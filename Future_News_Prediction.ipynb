```python
# Libray Imports

import requests
import json
import time
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
import spacy
from nltk import Tree
import time

# Function for Vader Sentiment Analysis


def sentiment_scores(sentence):
  
    sid_obj = SentimentIntensityAnalyzer()
  
    sentiment_dict = sid_obj.polarity_scores(sentence)
      
    if sentiment_dict['compound'] >= 0.05 :
        return "Positive"
  
    elif sentiment_dict['compound'] <= - 0.05 :
        return "Negetive"
  
    else :
        return "Neutral"


# Class for news and Article Details


class fetchNews():

  apikey = "3106a309c9mshfcf1f18d41d548fp1abc47jsn74cbf1cb3962"
  apihost = "contextualwebsearch-websearch-v1.p.rapidapi.com"

  def __init__(self,fromTime,toTime,pageSize=50):
    self.pageSize = str(pageSize)
    self.fromTime = str(fromTime)
    self.toTime = str(toTime)

  def getNews(self,searchTopic,pageNumber=1):

    newsArticle_Details = list()

    url = "https://contextualwebsearch-websearch-v1.p.rapidapi.com/api/search/NewsSearchAPI"
    querystring = {"q":searchTopic,"pageNumber":pageNumber,"pageSize":self.pageSize,"autoCorrect":"true","fromPublishedDate":self.fromTime,"toPublishedDate":self.toTime}

    headers = {
      'x-rapidapi-key': fetchNews.apikey,
      'x-rapidapi-host': fetchNews.apihost
    }

    response = requests.request("GET", url, headers=headers, params=querystring).json()['value']

    for eachRes in response:
      eachArticle = {}
      eachArticle['news_title'] = eachRes['title']
      eachArticle['news_description'] = eachRes['description']
      eachArticle['news_full_story'] = eachRes['body']
      eachArticle['news_snipet'] = eachRes['snippet'].replace("<b>","").replace("</b>","")
      eachArticle['news_sentiment'] = sentiment_scores(eachRes['snippet'].replace("<b>","").replace("</b>",""))
      eachArticle['news_timestamp'] =  eachRes['datePublished']
      eachArticle['news_url'] = eachRes['url']
      newsArticle_Details.append(eachArticle)

    return newsArticle_Details
```

```python
## Function for NLK Tree 

def tok_format(tok):
    return "_".join([tok.orth_, tok.pos_])

## Function for NLTK tree 

def to_nltk_tree(node):
    if node.n_lefts + node.n_rights > 0:
        return Tree(tok_format(node), [to_nltk_tree(child) for child in node.children])
    else:
        return tok_format(node)


## Main class for pos tag creation

class find_pos_tags():

  def __init__(self,text):
    ## Initial containers
    self.sent_list = list() ## Container for sentence list
    self.word_info = {} ## container for word info
    self.doc = en_nlp(text) ## container for tokenized docs

  def token_and_lemma(self): ## function for pos tagging and lemmatization
    for token in self.doc:
      self.word_info[token.text] = [token.pos_,token.lemma_]
    
    return True


  def recurse_tree(self,parent,child): ## Recursive function for action and intent detection
    if child.n_lefts + child.n_rights > 0:
      if (self.word_info[str(parent)][0] == 'VERB' and self.word_info[str(child)][0] == 'PROPN') or (self.word_info[str(parent)][0] == 'VERB' and self.word_info[str(child)][0] == 'NOUN'):
        self.sent_list.append((self.word_info[str(parent)][1],self.word_info[str(child)][1]))

      for chil in child.children:
        find_pos_tags.recurse_tree(self,child,chil)

    else:
      if (self.word_info[str(parent)][0] == 'VERB' and self.word_info[str(child)][0] == 'PROPN') or (self.word_info[str(parent)][0] == 'VERB' and self.word_info[str(child)][0] == 'NOUN'):
        self.sent_list.append((self.word_info[str(parent)][1],self.word_info[str(child)][1]))

  
  def conjugate_intent(self,intent): ## sub traversal function to add the nearby conjugate intents with proper noun childs
    for token in self.doc:
      if str(token) == intent:
        try:
          if (self.word_info[str(token.nbor())][0] == 'ADJ') or (self.word_info[str(token.nbor())][0] == 'PROPN') or (self.word_info[str(token.nbor())][0] == 'NOUN'):
            return intent+' '+str(token.nbor())
          else:
            pass

        except:
          pass

        try:
          if (self.word_info[str(token.nbor(-1))][0] == 'ADJ') or (self.word_info[str(token.nbor(-1))][0] == 'PROPN') or (self.word_info[str(token.nbor(-1))][0] == 'NOUN'):
            return str(token.nbor(-1))+' '+intent
          else:
            pass

        except:
          pass

      else:
        pass
        
  def add_more_intent(self): ## Main traversal function to conjugate intent
    interMididate_list = list()
    for eachEntry in self.sent_list:
      if find_pos_tags.conjugate_intent(self,eachEntry[1]) != None:
        interMididate_list.append((eachEntry[0],find_pos_tags.conjugate_intent(self,eachEntry[1])))
      else:
        interMididate_list.append((eachEntry[0],eachEntry[1]))

    self.sent_list = interMididate_list
  

  def get_extracted_details(self): ## function for pos tagged details extraction
    return (self.sent_list,self.word_info)
```

```python
### pos tag class example 

en_nlp = spacy.load('en_core_web_sm',disable=['ner','textcat'])
snippet = "Reuters, citing cyber intelligence firm Cypharma, said the two companies making the Corona vaccine in India had been targeted by Chinese government-backed hackers."
doc = en_nlp(snippet)
fp = find_pos_tags(snippet)
completed_lemma = fp.token_and_lemma()
for sent in doc.sents:
  for child in sent.root.children:
    fp.recurse_tree(sent.root,child)

fp.add_more_intent()
extracted_info = fp.get_extracted_details()
print(extracted_info[0])

```
